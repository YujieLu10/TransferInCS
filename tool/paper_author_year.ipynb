```python
for i in range (1, 13):
    print("start file {}".format(i))
    ori_papercitations = pd.read_csv('/data/pcs/split_papercitations/papercitations_{}.tsv'.format(i), delimiter=",|\t", usecols=[1,2], encoding='unicode_escape', header=None)
    ori_papercitations.columns = ['citingpaperid', 'citedpaperid']
    paper_citation_year = pd.merge(ori_papercitations, ori_paperyear, left_on='citingpaperid', right_on='paperid')
    try:
        paper_citation_year.to_csv('/data/pcs/split_papercitations/paper_citation_year_{}.tsv'.format(i), columns=['citingpaperid', 'citedpaperid', 'paperyear'])
    except:
        print("error")
    print("done file {}".format(i))
```
